```python
import numpy as np

x=['python', 'c', 'c++', 'java']
y=[85,70,60,82]
z=[10,20,30,40]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]


plt.xlabel('language', fontsize=20)
plt.ylabel('no', fontsize=20)
plt.title('Bar Graph', fontsize=20)

plt.bar(p,y, width, color='r',label='popularity') # x ki jga p
plt.bar(p1,z, width, color='g',label='popularity1')

plt.legend()
plt.show()
```

```python
import numpy as np

x=['python', 'c', 'c++', 'java']
y=[85,70,60,82]
z=[10,20,30,40]
width=0.2 # width ki waha se sai ho gaya


p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('language', fontsize=20)
plt.ylabel('no', fontsize=20)
plt.title('Bar Graph', fontsize=20)

plt.barh(p,y,width, color='r',label='popularity') # y walay horizontal
plt.barh(p1,z,width, color='g',label='popularity1') # x walay b horizontal

plt.xticks(p+width/2,x, rotation=20) 
plt.legend()
plt.show()
```
