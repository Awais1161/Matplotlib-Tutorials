```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y)

plt.savefig('line')

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y)

plt.savefig('line1', dpi=2000)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y)

plt.savefig('line3', dpi=100, facecolor='g')

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y)

plt.savefig('line4.pdf', dpi=100, facecolor='g')

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y)

plt.savefig('line4.jpg', dpi=100, facecolor='g')  # by default is PNG

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y)

plt.savefig('line11', dpi=100, facecolor='g', transparent=True)  

plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y)

plt.savefig('line22', dpi=100, facecolor='g', transparent=False, bbox_inches='tight')  
plt.tight_layout()
plt.show()
```
