```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,1,5,2,4]

plt.plot(x,y)

plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,1,5,2,4]

plt.plot(x,y)

plt.xticks(x)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,1,5,2,4]

plt.plot(x,y)

plt.xticks(x)
plt.yticks(y)
plt.show()
```

```python
x=[1,2,3,4]
y=[3,1,5,2]

plt.plot(x,y)

plt.xticks(x, labels=['python', 'c', 'c++', 'java'])
plt.yticks(y)
plt.show()
```
