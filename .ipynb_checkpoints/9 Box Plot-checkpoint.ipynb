```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x, notch=True) # false is default
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x, vert=False) # True is default
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x, widths=0.8) 
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x, labels=['python']) 
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x, labels=['python'], patch_artist=True) 
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x, labels=['python'], patch_artist=False) 
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x, labels=['python'], patch_artist=True, showmeans=True) 
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x, labels=['python'], showmeans=True) 
plt.show()
```

```python
x=[10,20,30,40,50,60,70, 120]
plt.boxplot(x, labels=['python'], showmeans=True,sym=' ')  # hide
plt.show()
```

```python
x=[10,20,30,40,50,60,70, 120]
plt.boxplot(x, labels=['python'], showmeans=True,capprops=dict(color='r')) 
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x, labels=['python'], showmeans=True,capprops=dict(color='r'),whiskerprops=dict(color='r')) 
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x, labels=['python'], showmeans=True,capprops=dict(color='r'),
            whiskerprops=dict(color='r'),flierprops=dict(markeredgecolor='y'))  
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
x1=[10,20,40,60,80,90,100]
plt.boxplot(x, labels=['python'], showmeans=True,capprops=dict(color='r'),
            whiskerprops=dict(color='r'),flierprops=dict(markeredgecolor='y')) 
plt.boxplot(x1, labels=['python'], showmeans=True,capprops=dict(color='r'),
            whiskerprops=dict(color='r'),flierprops=dict(markeredgecolor='y')) 
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
x1=[10,20,40,60,80,90,100]
y=[x,x1]
plt.boxplot(y,  showmeans=True,capprops=dict(color='r'),
            whiskerprops=dict(color='r'),flierprops=dict(markeredgecolor='y')) 
 
plt.show()
```

```python
x=[10,20,30,40,50,60,70]
x1=[10,20,40,60,80,90,100]
y=[x,x1]
plt.boxplot(y,labels=['python', 'java'] , showmeans=True,capprops=dict(color='r'),
            whiskerprops=dict(color='r'),flierprops=dict(markeredgecolor='y')) 
 
plt.show()
```
