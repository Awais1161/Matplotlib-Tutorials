```python
import matplotlib.pyplot as plt


x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.step(x,y)


plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,2,3,5,4]

plt.step(x,y)


plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,2,3,5,4]

plt.step(x,y)


plt.title('Python')
plt.xlabel(' x - axis')
plt.ylabel(' y - axis')
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,2,3,5,4]

plt.step(x,y, marker='o')


plt.title('Python')
plt.xlabel(' x - axis')
plt.ylabel(' y - axis')
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,2,3,5,4]

plt.step(x,y, color='r', marker='o')


plt.title('Python')
plt.xlabel(' x - axis')
plt.ylabel(' y - axis')
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,2,3,5,4]

plt.step(x,y, color='r', marker='o', ms=10) # marker size


plt.title('Python')
plt.xlabel(' x - axis')
plt.ylabel(' y - axis')
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,2,3,5,4]

plt.step(x,y, color='r', marker='o', ms=10, mfc='g') # marker face color


plt.title('Python')
plt.xlabel(' x - axis')
plt.ylabel(' y - axis')
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,2,3,5,4]

plt.step(x,y, color='r', marker='o', ms=10, mfc='g', label='Python') 


plt.title('Python')
plt.xlabel(' x - axis')
plt.ylabel(' y - axis')
plt.legend()
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5]
y=[3,2,3,5,4]

plt.step(x,y, color='r', marker='o', ms=10, mfc='g', label='Python') 


plt.title('Python')
plt.xlabel(' x - axis')
plt.ylabel(' y - axis')
plt.legend(loc=2)
plt.grid()
plt.show()
```
