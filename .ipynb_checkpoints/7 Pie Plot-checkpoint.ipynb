```python
import matplotlib.pyplot as plt
x=[10, 20, 30, 40]
y=['c', 'c++', 'python', 'java']

plt.pie(x)

plt.show()
```

```python
x=[10, 20, 30, 40]
y=['c', 'c++', 'python', 'java']
ex=[0.4,0.0,0.0,0.0]
c=['r', 'b', 'g', 'm']
plt.pie(x,labels=y, explode=ex, shadow=True, colors=c, autopct='%0.1f%%', counterclock=False,wedgeprops={'linewidth': 10})  
plt.show()   # shadow ki width
```

```python
x=[10, 20, 30, 40]
y=['c', 'c++', 'python', 'java']
ex=[0.4,0.0,0.0,0.0]
c=['r', 'b', 'g', 'm']
plt.pie(x,labels=y, explode=ex, shadow=True, colors=c, autopct='%0.1f%%', counterclock=False,wedgeprops={'linewidth': 10, 
                                                                                                        'edgecolor':'y'})  
plt.show()   
```

```python
x=[10, 20, 30, 40]
y=['c', 'c++', 'python', 'java']
ex=[0.4,0.0,0.0,0.0]
c=['r', 'b', 'g', 'm']
plt.pie(x,labels=y, explode=ex, shadow=True, colors=c, autopct='%0.1f%%', counterclock=False,wedgeprops={'linewidth': 10, 
                            'edgecolor':'y'}, center=(6,6))  
plt.show() 
```

```python
x=[10, 20, 30, 40]
y=['c', 'c++', 'python', 'java']
ex=[0.4,0.0,0.0,0.0]
c=['r', 'b', 'g', 'm']
plt.pie(x,labels=y, explode=ex, shadow=True, colors=c, autopct='%0.1f%%', counterclock=False,wedgeprops={'linewidth': 10, 
                            'edgecolor':'y'}, rotatelabels=True) # False by default 
plt.show()
```

```python
x=[10, 20, 30, 40]
y=['c', 'c++', 'python', 'java']
ex=[0.4,0.0,0.0,0.0]
c=['r', 'b', 'g', 'm']
plt.pie(x,labels=y, explode=ex, shadow=True, colors=c, autopct='%0.1f%%', counterclock=False,wedgeprops={'linewidth': 10, 
                            'edgecolor':'y'}, rotatelabels=True)
plt.title(' Pie Graph')
plt.show()
```

```python
x=[10, 20, 30, 40]
y=['c', 'c++', 'python', 'java']
ex=[0.4,0.0,0.0,0.0]
c=['r', 'b', 'g', 'm']
plt.pie(x,labels=y, explode=ex, shadow=True, colors=c, autopct='%0.1f%%', counterclock=False,wedgeprops={'linewidth': 10, 
                            'edgecolor':'y'}, rotatelabels=True)

plt.legend()
plt.title(' Pie Graph')
plt.show()
```

```python
x=[10, 20, 30, 40]
y=['c', 'c++', 'python', 'java']
ex=[0.4,0.0,0.0,0.0]
c=['r', 'b', 'g', 'm']
plt.pie(x,labels=y, explode=ex, shadow=True, colors=c, autopct='%0.1f%%', counterclock=False,wedgeprops={'linewidth': 10, 
                            'edgecolor':'y'}, rotatelabels=True)

plt.legend(loc=2) # 1 si by default
plt.title(' Pie Graph')
plt.show()
```
