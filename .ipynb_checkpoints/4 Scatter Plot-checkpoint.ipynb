```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.title('Scttar Plot', fontsize=20)
plt.xlabel('Day', fontsize=20)
plt.ylabel('No.', fontsize=20)
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]

colors=[10,49,30,29,56,20,30] 
sizes=[100,200,400,500,200,100,700] 
plt.scatter(x,y, c=colors,s=sizes, cmap='viridis')
n=plt.colorbar()
n.set_label('Color Bar')
plt.title('Scattar Plot', fontsize=20)
plt.xlabel('Day', fontsize=20)
plt.ylabel('No.', fontsize=20)
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
z=[1,3,5,7,9,3,5]

colors=[10,49,30,29,56,20,30] 
sizes=[100,200,400,500,200,100,700] 
plt.scatter(x,y, c=colors,s=sizes, cmap='viridis')
plt.scatter(x,z, color='r',s=sizes)
n=plt.colorbar()
n.set_label('Color Bar', fontsize=20)
plt.title('Scattar Plot', fontsize=20)
plt.xlabel('Day', fontsize=20)
plt.ylabel('No.', fontsize=20)
plt.show()
```
